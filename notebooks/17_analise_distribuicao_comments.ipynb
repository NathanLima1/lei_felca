```python
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python

videos = "complete_dataset.csv"
comments = "comments_shorts_processed.csv"

videos_df = pd.read_csv(videos)
comments_df = pd.read_csv(comments)

comments_df['cleanedText'] = comments_df['cleanedText'].astype(str).fillna('')
```

```python
contagem_df = comments_df.groupby(['videoId', 'isShorts']).size().reset_index(name='qtd_comentarios')

longos = contagem_df[contagem_df['isShorts'] == 0]['qtd_comentarios']
shorts = contagem_df[contagem_df['isShorts'] == 1]['qtd_comentarios']

stat, p_valor = mannwhitneyu(longos, shorts, alternative='greater')

print(f"P-valor (Amostra Coletada): {p_valor}")

if p_valor < 0.05:
    print("A diferença é estatisticamente significativa.")
else:
    print("A diferença pode ser obra do acaso.")
```

```python
mapeamento_shorts = comments_df[['videoId', 'isShorts']].drop_duplicates()

videos_completos = videos_df.merge(mapeamento_shorts, on='videoId', how='inner')

videos_completos['commentCount'] = pd.to_numeric(videos_completos['commentCount'], errors='coerce')
videos_limpos = videos_completos.dropna(subset=['commentCount'])

longos_oficial = videos_limpos[videos_limpos['isShorts'] == False]['commentCount']
shorts_oficial = videos_limpos[videos_limpos['isShorts'] == True]['commentCount']

stat_real, p_valor_real = mannwhitneyu(longos_oficial, shorts_oficial, alternative='greater')

print(f"P-valor (Contagem Oficial YouTube): {p_valor_real}")
```
