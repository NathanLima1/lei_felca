```python
import pandas as pd
```

```python
def select_sample(group):
    n = min(len(group), 150)
    return group.sample(n=n, random_state=42)
```

```python
df = pd.read_csv("comments_dataset.csv")

df['commentPublishedAt'] = pd.to_datetime(df['commentPublishedAt'])
df['year'] = df['commentPublishedAt'].dt.year
df['month'] = df['commentPublishedAt'].dt.month

df_sample_shorts = df[df["isShorts"] == 1].groupby(['year', 'month'], group_keys=False).apply(select_sample, include_groups=False)
df_sample_shorts = df_sample_shorts.reset_index(drop= True)
df_sample_shorts['evaluation'] = ""
df_sample_shorts.to_csv("comments_to_evaluate_shorts.csv", index=False)

df_sample_longos = df[df["isShorts"] == 0].groupby(['year', 'month'], group_keys=False).apply(select_sample, include_groups=False)
df_sample_longos = df_sample_longos.reset_index(drop= True)
df_sample_longos['evaluation'] = ""
df_sample_longos.to_csv("comments_to_evaluate_longos.csv", index=False)
```
